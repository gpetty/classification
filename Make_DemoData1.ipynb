```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 6000
a1 = np.random.normal(loc=0.0, scale=1.0, size=5*N)
a2 = np.random.normal(loc=0.0, scale=1.0, size=5*N)
a3 = np.random.normal(loc=0.0, scale=1.0, size=10*N)
a4 = np.random.normal(loc=0.0, scale=1.0, size=10*N)
a5 = np.random.normal(loc=0.0, scale=1.0, size=10*N)
a6 = np.random.normal(loc=0.0, scale=1.0, size=10*N)
a7 = np.random.normal(loc=0.0, scale=1.0, size=5*N)
a8 = np.random.normal(loc=0.0, scale=1.0, size=5*N)

x1 = 3*a1
y1 = 0.5*a1**2 + a2

x2 = 2 + 2.0*a3
y2 = a4 - 3

x3 = a5 - 5.
y3 = 2.0*a6 + 2.

x4 = 1.2*a7 - 2.
y4 = 0.4*a8 + 5
```

```python
x = np.concatenate([x1,x2,x3,x4]).reshape((-1, 1))
y = np.concatenate([y1,y2,y3,y4]).reshape((-1, 1))

xy = np.concatenate([x,y], axis=1)
print(xy.shape)

np.random.shuffle(xy)

np.save(open('PSP_DemoData1', 'wb'), xy)
```

```python
x = xy[:3000,0]
y = xy[:3000,1]

npoints = len(x)

fig = plt.figure(figsize=(7,7))

s = 5
plt.scatter(x, y, s=s, c='k')
plt.annotate('$N_{total} = $%d' % npoints , (-9,-9), fontsize=15)

plt.xlim([-10,10])
plt.ylim([-10,10])

axes=plt.gca()
axes.set_aspect('equal')
plt.xticks([])
plt.yticks([])
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('Fig2.pdf')
plt.show()
```
